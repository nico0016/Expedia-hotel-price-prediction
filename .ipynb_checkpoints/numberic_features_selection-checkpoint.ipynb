# Part1: Split data into training, validation and test set manually

Codes below are used to split Expedia data into training, validation and test set.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

from scipy.stats import skew
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
print('training data has %d records'%len(train))
print('test data has %d records'%len(test))
```

```python
cols_train_only = [col for col in train.columns.unique().tolist() if col not in test.columns.unique().tolist()]
print('Columns only available in training data:',cols_train_only)
```

```python
train = train.drop(columns = cols_train_only)
train.head()
```

```python
# combine train and test data
all_data = pd.concat([train, test], ignore_index=True)
all_data.head()
```

```python
print('Whole dataset has %d records' % len(all_data))
```

```python
# remove the columns which have >50% NA
morethanhalfna_columns = ['visitor_hist_starrating','visitor_hist_adr_usd','srch_query_affinity_score','comp1_rate','comp1_inv','comp1_rate_percent_diff','comp2_rate','comp2_inv','comp2_rate_percent_diff','comp3_rate','comp3_inv','comp3_rate_percent_diff','comp4_rate','comp4_inv','comp4_rate_percent_diff','comp5_rate','comp5_inv','comp5_rate_percent_diff','comp6_rate','comp6_inv','comp6_rate_percent_diff','comp7_rate','comp7_inv','comp7_rate_percent_diff','comp8_rate','comp8_inv','comp8_rate_percent_diff']
X_train_lessna = X_train[[col for col in X_train.columns.unique().tolist() if col not in morethanhalfna_columns]]
X_train_lessna.columns
# remove the columns which present ID
id_feature_columns = ['srch_id','site_id','visitor_location_country_id','prop_country_id','prop_id','srch_destination_id']
X_train_lessna_noid = X_train_lessna[[col for col in X_train_lessna.columns.unique().tolist() if col not in id_feature_columns]]
X_train_lessna_noid.columns
# remove the boolean columns and date_time
boolean_columns = ['prop_brand_bool','srch_saturday_night_bool','random_bool','date_time']
X_train_numeric = X_train_lessna_noid[[col for col in X_train_lessna_noid.columns.unique().tolist() if col not in boolean_columns]]
X_train_numeric.columns

```
