```python
# TODO: format data by property --> we need to predict price for each property
property_price_data = data[['date_time','prop_id','price_usd']]
property_price_data[['prop_id','price_usd']].groupby('prop_id').mean()
```

```python
data['date_time'] = pd.to_datetime(data.date_time)
price_data = data[['date_time','price_usd']]
price_data.head()
```

```python
# create a "month" column that lists "month" signal
price_data['day'] = price_data['date_time'].apply(lambda x: x.day)
price_data['month'] = price_data['date_time'].apply(lambda x: x.month)
price_data['quarter'] = price_data['date_time'].apply(lambda x: x.quarter)
price_data.head()
```

```python
# compute variance across month, quarter, day
# using mean as comparison metric
price_data[['price_usd','month']].groupby('month').mean()
# can tell that average hotel price vary a lot across months
```

```python
# group by quarter
price_data[['price_usd','quarter']].groupby('quarter').mean()
```

```python
# group by day
price_data[['price_usd','day']].groupby('day').mean()
```
